```python
# importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
# loading the dataset
retail=pd.read_csv("C:/Users/praneethaa m/Downloads/SampleSuperstore.csv")
retail
```

```python
#Basic data exploration 
retail.shape
retail.info()
retail.describe()
```

```python
#handling the missing values 
retail.isnull().sum()
#no missing va;ues are found
```

```python
#check for duplicate data
duplicate=retail.duplicated()
print(duplicate.sum())
retail[duplicate]
```

```python
# droping the duplicate value rows
retail.drop_duplicates(inplace=True)
#cheaking for duplicate data again
dupl=retail.duplicated()
dupl.sum()
```

```python
#Drop unwanted columns 
retail=retail.drop(columns=['Ship Mode','Country','Postal Code','Segment'])
retail.head()
```

```python
#finding sales for each category 
sales_retail = retail.groupby('Category', as_index=False)['Sales'].sum()
subcat_retail = retail.groupby(['Category','Sub-Category'])['Sales'].sum()
subcat_retail['Sales']=map(int,subcat_retail)
sales_retail
```

```python
#visualizing the above sales data
fig,ax = plt.subplots(figsize=(10,5))
ax.bar(sales_retail['Category'],sales_retail['Sales'],color='pink',edgecolor='#1b470a',width=0.3) 
ax.set_title(label="Sales for each Category and sub-category", loc='center', pad=None)
ax.set_ylabel('Sales')
ax.set_xlabel('Category')
```

```python
# To display the profits for each state
profit_retail = retail.groupby('State', as_index=False)['Profit'].sum()
profit_retail.head()
```

```python
# going to plot a graph for displaying the overall profit or loss of every state 
fig,ax = plt.subplots(figsize=(10,6))
ax.bar(profit_retail['State'],profit_retail['Profit'],color='pink',edgecolor='#de124c') 
ax.set_title(label="Total Profit for each State", loc='center', pad=None)
ax.set_ylabel('Profit')
ax.set_xlabel('State')
ax.set_xticklabels(profit_retail['State'],rotation=90)
plt.tight_layout();
```

```python
# now lets check the relationship between profit and discount by using a boxplot
props = dict(boxes="pink", medians="Black", caps="black")
retail.boxplot(by='Discount', column='Profit',showfliers=False,figsize=(10,6),patch_artist=True,color=props)
```

```python
# AS OF NOW WE HAVE FINISHED THE PROFIT PART NOW LETS GO TO THE LOSS PART
#Lets calculate the loss based on the state
state_neg=profit_retail.loc[profit_retail.Profit < 0]
state_neg
fig,ax = plt.subplots(figsize=(10,5))
def plot_state(ax):
    ax.bar(state_neg['State'],state_neg['Profit'],color='pink',edgecolor='black') 
    ax.set_title(label="States which faced Loss", loc='center', pad=None)
    ax.set_ylabel('Loss')
    ax.set_xlabel('States')
plot_state(ax)
plt.tight_layout()

#thus from the below chart we infer that the states which has to concentrate more on thier profits rates are TEXAS,OHIO,PENNSYLVANIA,ILLINIOS
```

```python
#Loss calculation basend on region
state_neg2=retail.loc[retail.Profit < 0].reset_index(drop=True)
state_pos2=retail.loc[retail.Profit >= 0].reset_index(drop=True)
dft=state_neg2.groupby(['Region','State'], as_index=False)['Profit'].count()
dft=dft[dft.Profit > 40]
dft
fig,ax = plt.subplots(figsize=(10,6))
def plot_region(ax):
    d={'Central':'#FF69B4','East':'#FF1493','South':'#DB7093','West':'#C71585'}
    dft['color'] = dft['Region'].map(d)
    sns.barplot(data=dft, x=dft.State, y='Profit', hue='Region', palette=d, dodge=False, edgecolor='black')
    ax.set_xticklabels(dft['State'],rotation=90)
    ax.set_title(label=" Total Number of items faced loss in each state Grouped by region", loc='center', pad=None)
    ax.set_ylabel('Count')
    ax.set_xlabel('State')
    ax.legend(loc=1, fontsize='large',edgecolor='black', borderpad=1.0, title="Region", shadow=True)
    plt.show()
plot_region(ax)
plt.tight_layout()
```

```python
# LOSS DISTRIBUTION BASED ON CATEGORY
state_pos2['Discount'].value_counts().reindex(retail.Discount.unique(), fill_value=0).sort_values()
state_neg2['Discount'].value_counts().reindex(retail.Discount.unique(), fill_value=0)
state_pos2.head()
fig,ax = plt.subplots(figsize=(10,5))
def plot_cat(ax):
    dftn=state_neg2.groupby('Category', as_index=False)['Profit'].count()
    ax.bar(dftn['Category'],dftn['Profit'],color='pink',edgecolor='black',width=0.3)
    ax.set_title(label="Total number of items in each category which faced loss", loc='center', pad=None)
    ax.set_ylabel('Count')
    ax.set_xlabel('Category')
plot_cat(ax)
plt.show()

### FROM THE BELOW GRAPH WE INFER THAT THE DISCOUNT GIVEN FOR OFFICE SUPPLIES IS BEING HIGH AND THAT MAY BE THE REASON WHY I HAS FACED A HUGE LOSS IN PROFIT.
```

```python
fig,ax = plt.subplots(figsize=(10,5))
def plot_disccat(ax):
    sns.scatterplot(x="Discount", y="Profit", data=retail, hue="Category",ax=ax,s=200,palette='YlOrRd',edgecolor='brown')
    ax.legend(loc=1, fontsize='large',edgecolor='black', borderpad=1.0, shadow=True)
    ax.set_title('Profit achieved by each Category for each Discount value.')
plot_disccat(ax)
plt.show()
```
